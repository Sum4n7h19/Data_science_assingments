```python
import seaborn as sns
import plotly.graph_objects as px
import plotly.express as pt
import plotly.figure_factory as fac
```

```python
#Q1
titanic=sns.load_dataset('titanic')
```

```python
fig=px.Figure()
fig.add_trace(px.Scatter(x=titanic.age,y=titanic.fare,mode='markers'))
```

```python
#Q2
tips=sns.load_dataset('tips')
```

```python
fig=pt.box(tips ,x='day',y='total_bill',color='sex')
fig.show()
```

```python
#Q3
fig=pt.histogram(tips ,x='sex',y='total_bill',color='day',barmode='group',
           facet_col='smoker',pattern_shape='smoker')
fig.show()
```

```python
#Q4
iris=sns.load_dataset('iris')
```

```python
fig=pt.scatter_matrix(iris,dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],color='species')
fig.show()
```

```python
#Q5
g1=iris.sepal_length
g2=iris.sepal_width
g3=iris.petal_length
g4=iris.petal_width
data=[g1,g2,g3,g4]
group_labels=["sepal_length", "sepal_width", "petal_length", "petal_width"]
fig = fac.create_distplot(data,group_labels)
fig.show()
```

Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

1. **Scatter Plot:**
   - Use: Scatter plots are used to visualize the relationship between two numeric variables. Each data point is represented as a point on the graph, with one variable on the x-axis and the other on the y-axis. It helps in identifying patterns, clusters, or correlations between variables.

2. **Bar Plot:**
   - Use: Bar plots display categorical data with rectangular bars, where the length of each bar represents the value of the corresponding category. They are used to compare the values of different categories and show the distribution of categorical data.

3. **Histogram:**
   - Use: Histograms are used to represent the distribution of a single numeric variable. It divides the data into discrete bins and displays the frequency or count of data points within each bin. Histograms help visualize data distribution and identify patterns such as skewness or outliers.

4. **Heatmap:**
   - Use: Heatmaps are graphical representations of data using a color-coded matrix. They are useful for displaying the correlation between multiple variables in a dataset. Heatmaps are commonly used in exploratory data analysis and in identifying patterns in large datasets.

5. **Line Plot:**
   - Use: Line plots are used to visualize the trend or relationship between two numeric variables over a continuous interval. They are helpful in showing how a variable changes over time or across different levels of another variable.
   

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Q2
fmri=sns.load_dataset('fmri')
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal)
```

```python
#Q3
titanic=sns.load_dataset('titanic')
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=titanic.pclass,y=titanic.age)
plt.title("Age vs pclass")

plt.figure(figsize=(10,6))
plt.subplot=(1,2,2)
sns.boxplot(x=titanic.pclass,y=titanic.fare)
plt.title("Fare vs pclass")

plt.show()
```

```python
#Q4
diamond=sns.load_dataset("diamonds")
```

```python
sns.displot(x=diamond.price, data=diamond,hue='cut')
```

```python
#Q5
iris=sns.load_dataset('iris')
```

```python
sns.pairplot(iris,hue='species')
```

```python
#Q6
flights=sns.load_dataset('flights')
flights_pivot = flights.pivot('month', 'year', 'passengers')
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights_pivot,fmt='d',annot=True)
plt.title('Flights Heatmap')
```

```python
#Q1
import scipy.stats as stats

# Sample data
mean = 50
std_dev = 5
sample_size = 100  # Replace with your actual sample size

# Calculate standard error of the mean
sem = std_dev / (sample_size**0.5)

# Specify the confidence level
confidence_level = 0.95

# Calculate the margin of error
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * sem

# Calculate the confidence interval
lower_bound = mean - margin_of_error
upper_bound = mean + margin_of_error

# Print the confidence interval
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#Q2
import scipy.stats as stats
import numpy as np

#Assume the total no of m&m's are 201

# Observed frequencies
observed = np.array([40, 40, 40, 21, 20, 40])

# Expected frequencies based on the given distribution
expected = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(observed)

# Perform the chi-square test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Compare p-value to significance level
alpha = 0.05
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis. The distribution of M&M colors is different from the expected distribution.")
else:
    print("Fail to reject the null hypothesis. The distribution of M&M colors matches the expected distribution.")
```

```python
#Q6
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2

# Degrees of freedom
df = 10

# Chi-square statistic to shade
chi_statistic = 15

# Values for the x-axis
x = np.linspace(0, 30, 1000)

# Chi-square distribution
y = chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, y, label=f'Chi-Square Distribution (df={df})')

# Shade the area corresponding to the chi-square statistic
x_shaded = np.linspace(chi_statistic, 30, 1000)
y_shaded = chi2.pdf(x_shaded, df)
plt.fill_between(x_shaded, y_shaded, color='red', alpha=0.3, label=f'Shaded Area (>= {chi_statistic})')

# Label the axes
plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density Function (PDF)')
plt.title('Chi-Square Distribution with Shaded Area')
plt.legend()

# Show the plot
plt.show()
```

```python
#Q8
import numpy as np
from scipy.stats import chisquare

# Given data
observed_tails = 45
total_flips = 100

# Expected frequencies for a fair coin
expected_tails = total_flips / 2
expected_heads = total_flips / 2

# Observed frequencies
observed_frequencies = np.array([observed_tails, total_flips - observed_tails])
expected_frequencies = np.array([expected_tails, expected_heads])

# Perform chi-square goodness of fit test
chi2_stat, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

# Print the results
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Check significance level
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The observed frequencies do not match the expected frequencies of a fair coin.")
else:
    print("Fail to reject the null hypothesis. The observed frequencies match the expected frequencies of a fair coin.")
```

```python
#Q11
import scipy.stats as stats

# Given data
sample_mean = 72
sample_std_dev = 10
population_mean_h0 = 70
sample_size = 30
significance_level = 0.05

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean_h0) / (sample_std_dev / (sample_size ** 0.5))

# Calculate the degrees of freedom
degrees_of_freedom = sample_size - 1

# Calculate the critical t-value for a two-tailed test
critical_t_value = stats.t.ppf(1 - significance_level / 2, df=degrees_of_freedom)

# Perform the hypothesis test
p_value = 2 * (1 - stats.t.cdf(abs(t_statistic), df=degrees_of_freedom))

# Print the results
print(f"T-Statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Check significance level
if p_value < significance_level:
    print("Reject the null hypothesis. There is sufficient evidence to suggest that the population mean is different from 70.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest that the population mean is different from 70.")
```
